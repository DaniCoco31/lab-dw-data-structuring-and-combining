2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
df_mkt.isna()
```

```python
df_mkt.isnull()
```

```python
#the average customer lifetime value per gender and education level.
pivot_table = df_mkt.pivot_table(
    values='customer_lifetime_value',
    index=['gender', 'education'],
    aggfunc='mean'  # Calculate the average for customer_lifetime_value
)

print(pivot_table)
```

```python
pivot_table.dtypes
highest_average = pivot_table.idxmax()
lowest_average = pivot_table.idxmin()
print(f'Highest average: {highest_average}\n')
print(f'Lowest average: {lowest_average}\n')
print('The highest average is in the women with education of High School \nor below and the lowest average resides within the woman with a Doctor degree. ')
```
